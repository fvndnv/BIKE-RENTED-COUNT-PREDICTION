```python
def plot_by_hour(data, year=None, agg='sum'):
    dd=data
    if year:
        dd=dd[dd.datetime.dt.year==year]
    dd.loc[:,('hour')]=dd.datetime.dt.hour
    by_hour=dd.groupby(['hour', 'workingday'])['count'].agg(agg).unstack()
    by_hour.plot(kind='bar', ylim=(0,80000), figsize=(15,10), width=0.9, title="year= {0}".format(year))
```

```python
plot_by_hour(train,2011)
```

```python
def bartplot(data, column1, column2):
    x=data[column1]
    y=data[column2]
    plt.bar(x, y)
```

```python
bartplot(train, 'holiday', 'count')
```
